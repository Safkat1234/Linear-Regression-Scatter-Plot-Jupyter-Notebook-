```python
%matplotlib inline
from pylab import*
import numpy as np
np.random.seed(4)

x=np.random.normal(3.0,1.0,1000)
y=np.random.normal(50.0,10.0,1000)/x

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

a=np.array(x)
b=np.array(y)

p9=np.poly1d(np.polyfit(a,b,9))

xp=np.linspace(0,6,300)
plt.scatter(x,y)
plt.plot(xp,p9(xp),c='b')
plt.show()
```

```python
#Gradient Descent || Calculating Slope || Update weights
import numpy as np

weights=np.array([1,2])
inputs=np.array([4,5])

target=8
learning_rate=0.01

pred=(weights*inputs).sum()
error=pred-target
print(error)
```

```python
gradient=2*inputs*error
weights_u=weights-learning_rate*gradient
pred=(weights_u*inputs).sum()

error_u=pred-target
print(error_u)
```
